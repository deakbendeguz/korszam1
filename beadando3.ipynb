```python
import tensorflow as tf
```

```python
x1 = tf.random.uniform([32,10,100,3],minval=-10,maxval=10,dtype=float)
x2 = tf.random.uniform([32,10,100,3],minval=-10,maxval=10,dtype=float)
print(x1)
```

```python
tavolsag1 = tf.math.reduce_euclidean_norm(
    x1-x2,axis = -1)
print(tavolsag1) #először erre gondoltam de itt a dimenzió nem megfelelő
```

```python
def lastdimdist(X,Y):
    Xnew = tf.expand_dims(X, axis=3) 
    Ynew = tf.expand_dims(Y, axis=2) 
    tavolsag = tf.math.reduce_euclidean_norm(
    Xnew-Ynew,axis=-1
    )
    return tavolsag
    #itt a tenzorokat expendelve a megfelelő indexek egymásból vonódnak ki
```

```python
print(lastdimdist(x1,x2))
```
